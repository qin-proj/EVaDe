```python
import matplotlib.pyplot as plt

# Assuming n_df is a predefined DataFrame with the necessary data
Dsp_V = list(n_df['Dsp/V'])
Dsam_V = list(n_df['Dsam/V'])

# Separate the data into two lists for "key genes" and "other genes"
key_genes_Dsp_V = [Dsp for Dsp, Dsam in zip(Dsp_V, Dsam_V) if Dsp > 845 and Dsam < 13.34]
key_genes_Dsam_V = [Dsam for Dsp, Dsam in zip(Dsp_V, Dsam_V) if Dsp > 845 and Dsam < 13.34]
other_genes_Dsp_V = [Dsp for Dsp, Dsam in zip(Dsp_V, Dsam_V) if not (Dsp > 845 and Dsam < 13.34)]
other_genes_Dsam_V = [Dsam for Dsp, Dsam in zip(Dsp_V, Dsam_V) if not (Dsp > 845 and Dsam < 13.34)]

plt.figure(figsize=(8, 8))

# Plot "key genes" in red
plt.scatter(key_genes_Dsp_V, key_genes_Dsam_V, s=5, c='r', alpha=0.6, label='Key genes')

# Plot "other genes" in black
plt.scatter(other_genes_Dsp_V, other_genes_Dsam_V, s=5, c='k', alpha=0.4, label='Other genes')

plt.xlabel('$D_{sp}^{n}$ / $V_{human}^{n}$', fontsize=18,labelpad=13,fontname='Arial')
plt.ylabel('$D_{sam}^{n}$ / $V_{human}^{n}$', fontsize=18,labelpad=8,fontname='Arial')
plt.xlim(0, 5000)
plt.ylim(0, 400)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
# Add the legend to the plot
plt.legend(loc='upper right',fontsize=18,markerscale=4)

plt.show()
```
